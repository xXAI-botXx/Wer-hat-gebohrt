```python
import os

import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

import sys
sys.path.append('../src')

train_df = pd.read_csv("../../data/prepared_train_data_2021-11-02.csv")
test_df = pd.read_csv("../../data/prepared_test_data_2021-11-02.csv")
pd.set_option('display.max_columns', 500)
train_df
```

```python
#Current__mean --> Stromstrke im durchschnitt
#Voltage__minimum, Voltage__sum_values, Voltage__count_below_mean -- Sind das die richtigen Volt angaben ?

# Berechnung vom Widerstand Ohm
train_df["Resistance"] = train_df["Current__mean"] / train_df["Voltage__minimum"]

# Berechnung der Elektrischen Leistung P
train_df["Power"] = train_df["Current__mean"] * train_df["Voltage__minimum"]

# Berechnung der Arbeit W --- Wo ist die Zeit ?
#train_df["Work"] = train_df["Current__mean"] * train_df["Voltage__minimum"] * train_df["Time"]

# Berechnung von mAh(milli Ampere Stunden)
train_df["mAh"] = (train_df["Current__mean"] / 60) * 1000

# Berechnung von Wh(Watt Stunden)
train_df["Wh"] = train_df["mAh"] * train_df["Voltage__minimum"]

train_df
```
